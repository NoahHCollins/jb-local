
## Outline  <a class="tocSkip">

### Part 1: Stress and Tensors
1. Cauchy Stress Tensor
2. (Stress) tensor symmetry 
3. Coordinate transformation (stress) tensors
4. Shear and normal stresses
5. Tensor invariants

## Learning outcomes  <a class="tocSkip">

1. Understand the meaning of different components of 3D Cauchy stress tensor
2. Know how to determine state of stress on given plane
3. Be able to decompose a rank 2 tensor into symmetric and anti-symmetric components
4. Be able to transform rank 2 tensor to a new basis
5. Be able to determine shear and normal stresses on a plane

## Cauchy Stress

The Cauchy stress tensor is a second-order tensor that completely defines the state of stress at a point inside a material in its deformed state. It is used for the stress analysis of materials undergoing small deformations, whereas for large deformations other methods have to be used such as the Piola-Kirchhoff stress tensor. 

```{figure} Images/Lecture2/Fig1.png
:width: 30%
``` 

By considering the traction of the stress vector, we can see that: 
 
$$ 
\large{\mathbf{t_1} = \mathbf{t(\hat{n}_1)} = \lim_{\Delta A \to 0} \frac{\Delta \mathbf{f}}{\Delta A_1}} 
$$
 
$$ 
\large{\mathbf{t_1} = (\sigma_{11},\sigma_{12},\sigma_{13})}
$$
    
As such, to fully describe the stress in all directions, nine components are requried. 

$$
\large{\sigma_{11}, \sigma_{12}, \sigma_{13} \hspace{0.1cm} \text{for} \hspace{0.1cm} \Delta A_1} 
$$

$$
\large{\sigma_{22}, \sigma_{21}, \sigma_{23} \hspace{0.1cm} \text{for} \hspace{0.1cm} \Delta A_2} 
$$

$$
\large{\sigma_{33}, \sigma_{31}, \sigma_{32} \hspace{0.1cm} \text{for} \hspace{0.1cm} \Delta A_3} 
$$

    
The first index (j) describes the orientation of the plane, whereas the second index (i) describes the orientation of the force. However _are nine components sufficient?_

## Plane area as a vector <a name = "Plane"></a>

The area of plane _S_ can be defined in terms of vectors, assuming that $S_0$ and $\theta$ are known. 

$$
\mathbf{S} = S \mathbf{\hat{n}}
$$

$$
S_0 = \mathbf{S} \cdot \mathbf{\hat{n}_0} = S\mathbf{\hat{n}} \cdot \mathbf{\hat{n}_0} = S\cos{\theta}
$$

$$
\Rightarrow S = \frac{S_0}{\cos{\theta}}
$$


```{figure} Images/Lecture2/Fig2.png
:width: 30%
``` 

### Example
Let us now try to answer whether nine components are sufficient to capture all the stresses acting on an object.
<br>
We can demonstrate this by taking the equilibrium for a tetrahedron. 

```{figure} Images/Lecture2/Fig3.png
:width: 40%
``` 

Given the stresses acting on surfaces $A_1, A_2 \; \text{and} \; A_3$.
Find $\mathbf{t(\hat{n})}. $

```{figure} Images/Lecture2/Fig4.png
:width: 40%
``` 


$$
\text{1.)} \hspace{0.2cm} \mathbf{\hat{n}} = - \mathbf{\hat{x}_1} \;\; \Rightarrow \Delta A_1 = \Delta A \cos{\theta_1} 
$$

$$
\text{2.)} \hspace{0.2cm} \mathbf{\hat{n}} = - \mathbf{\hat{x}_2} \;\; \Rightarrow \Delta A_2 = \Delta A \cos{\theta_2} 
$$

$$
\text{3.)} \hspace{0.2cm} \mathbf{\hat{n}} = - \mathbf{\hat{x}_3} \;\; \Rightarrow \Delta A_3 = \Delta A \cos{\theta_3} 
$$ 

$$
\text{4.)} \hspace{0.2cm} \mathbf{\hat{n}} = - (n_1, n_2, n_3) \;\; \Rightarrow n_i = \cos{\theta_i}, \Delta A_4 = \Delta A 
$$


Taking equilibrium gives:

$$
\sum f_1 = t_1 \Delta A - \sigma_{11} \Delta A \cos{\theta_1} - \sigma_{21} \Delta A \cos{\theta_2} = \sigma_{31} \Delta A \cos{\theta_3} = 0 
$$


This gives : 


$$
t_1 = \sigma_{11} n_1 + \sigma_{21} n_2 + \sigma_{31} n_3 
$$

and similarly 

$$
t_2 = \sigma_{12} n_1 + \sigma_{22} n_2 + \sigma_{32} n_3 
$$

$$
t_3 = \sigma_{13} n_1 + \sigma_{23} n_2 + \sigma_{33} n_3 
$$

```{figure} Images/Lecture2/Fig5.png
:width: 40%
``` 

$$
\large{\Rightarrow \color{blue}{t_i = \sigma_{ji} n_j}} 
$$

This is known as the Einstein convention. 

## 3-D Stress State <a name = "3D"></a>

```{figure} /Chapter2/Images/Lecture2/Fig6.png
:width: 50%
```  

The first index represents the orientiation of the plane whereas the second index represents the orientation of the force. 
<br>
The force is considered positive in the direction of the normal (as shown). 

$$
t_i = \sigma_{ji} n_j \hspace{4cm} \mathbf{t} = \sigma^T \cdot \mathbf{\hat{n}} \hspace{2cm} \text{Transpose} \hspace{0.2cm} \sigma_{ji} = \sigma_{ij}^T
$$

_Note the unusual index order, j comes first, not i_

In matrix notation, this gives: 

$$
\large{\mathbf{t} = 
\begin{bmatrix}
\sigma_{11} & \sigma_{21} & \sigma_{31} \\
\sigma_{12} & \sigma_{22} & \sigma_{32} \\
\sigma_{13} & \sigma_{23} & \sigma_{33} \\
\end{bmatrix}}
\cdot \mathbf{\hat{n}}
$$

### Important Points
1. __t__ and $\mathbf{\hat{n}}$ are tensors of rank 1 (i.e vectors) and $\sigma$ is a tensor of rank 2 in 3-D. 
2. Compressive stresses have a negative sign whereas tensile forces have a positive sign. 
3. The diagonal members of the matrix (i.e when i = j) are normal stresses and the remaining members are shear stresses.

__Rank 2 tensors__ can be written as __square matrices__ and have algebraic properties similar to matrices. 

### Tensor Symmetry <a name = "TensorSym"></a>

A tensor can be symmetric in 1 or more indices. 

For example, a rank 2 tensor is symmetric if: $S_{ij} = S_{ji} \Rightarrow \mathbf{S} = \mathbf{S^T}$, 
and vice versa, a rank 2 tensor is anti-symmetric if: $S_{ij} = - S_{ji} \Rightarrow \mathbf{S} = - \mathbf{S^T}$.

For tensors of higher rank: $S_{ijk} = S_{jik}$ for all i,j,k $\Rightarrow$ symmetric in __i,j__.

For an antisymmetric $\mathbf{T}$ of rank 2 (with $T_{ij} = 0, \text{for } i = j$), the number of independent components is given by : $ \large{\frac{n(n-1)}{2}} $ 
<br>
For a symmetric $\mathbf{T}$ of rank 2, the number of independent components is given by : $ \large{\frac{n(n+1)}{2}} $ 
<br>
where $n$ is the dimension of the tensor. 

Any tensor $T$ of rank 2 can be decomposed into its symmetric and antisymmetric parts: 

$$
\large{\mathbf{T}_{ij} = \frac{\mathbf{T}_{ij} + \mathbf{T}_{ji}}{2} + \frac{\mathbf{T}_{ij} - \mathbf{T}_{ji}}{2}}
$$

### Symmetry of the Stress Tensor  <a name = "Symmetry"></a> 

If we consider an cube with dimensions $x_1$, $x_1$ and $x_3$ with stresses $\sigma_{12}$ and $\sigma_{21}$ acting on its surface. 

```{figure} /Chapter2/Images/Lecture2/Fig7.png
:width: 40%
```  

We can see from the figure above that the following stress couple forms in the $\large{x_2}$-$\large{x_3}$ plane with moment arm $\large{\frac{\Delta x_1}{2}}$. 

```{figure} /Chapter2/Images/Lecture2/Fig8.png
:width: 25%
``` 

Similarly, another stress couple forms in the $x_1$ $x_3$ plane with moment arm $\frac{\Delta x_2}{2}$. 

```{figure} /Chapter2/Images/Lecture2/Fig8.png
:width: 25%
``` 

Assuming static equilibrium, we can take a moment balance in the $x_3$ direction. This gives: 

$$
m_3 = [\sigma_{12} + (\sigma_{12} + \Delta x_1 \frac{\partial \sigma_{12}}{\partial x_1})] \Delta x_2 \Delta x_3 \cdot \frac{\Delta x_1}{2} - [\sigma_{21} + (\sigma_{21} + \Delta x_2 \frac{\partial \sigma_{21}}{\partial x_2})] \Delta x_1 \Delta x_3 \cdot \frac{\Delta x_2}{2}
$$
    
$ \hspace{7.2cm} \color{blue}{\text{stress couple 1}} \hspace{1.1cm} \color{blue}{\text{area 1}} \hspace{0.4cm} \color{blue}{\text{arm 1}} \hspace{1.4cm} \color{red}{\text{stress couple 2}} \hspace{0.8cm}  \color{red}{\text{area 2}} \hspace{0.6cm} \color{red}{\text{arm 2}} $

$$
\Rightarrow [2\sigma_{12} + \Delta x_1 \frac{\partial \sigma_{12}}{\partial x_1}] - [2\sigma_{21} + \Delta x_2 \frac{\partial \sigma_{21}}{\partial x_2}] = 0
$$
    
$$
\lim_{\Delta x_1, \Delta x_2 \to 0} \Rightarrow \sigma_{12} = \sigma_{21}
$$

Balancing $m_1$ and $m_2$: 

$$
\large{\sigma_{23} = \sigma_{32}} \hspace{1cm} \text{and} \hspace{1cm} \large{\sigma_{13} = \sigma_{31}}
$$
 
$$ 
\mathbf{t = \sigma^{T} \cdot \hat{n} \Rightarrow t = \sigma \cdot \hat{n}} 
$$

Thus, the stress tensor is __symmetric__.

## Transformations of Tensors <a name = "Transformations"></a>

Recap from Lecture 1.2 that the physical parameters should not depend on the coordinate frame. __Tensors follow the same linear transformation laws__. 

For vectors on an orthonormal basis: 

$$
\mathbf{v'} = \mathbf{A} \mathbf{v}
$$

$$
\Rightarrow \mathbf{v'} = 
\begin{bmatrix}
\alpha_{11} & \alpha_{12} \\
\alpha_{21} & \alpha_{22} \\
\end{bmatrix}
\mathbf{v}
$$


Coefficients $\alpha {ij}$ depend on the angle $\phi$ between $x_1$ and $ x'_1$ (or $x_2$ and $ x'_2$). 
<br>

```{figure} Images/Lecture2/Fig10.PNG
:width: 25%
```  

$$
\mathbf{v'} =
\begin{bmatrix} 
\cos{\phi}  & \sin{\phi} \\
-\sin{\phi}  & \cos{\phi} \\
\end{bmatrix} 
\mathbf{v}
= 
\begin{bmatrix} 
\cos{\phi}  & \cos{90 - \phi} \\
\cos{90 + \phi}  & \cos{\phi} \\
\end{bmatrix} 
\mathbf{v}
$$

$$
\color{blue}{\large{\alpha_{ij} = \mathbf{\hat{e}'_i} \cdot \mathbf{\hat{e}_j}}}
$$

$$
\text{Inverse Transform} \hspace{0.5cm} \text{v}_j = \alpha_{ji} \text{v'}_j \hspace{1.2cm}\alpha_{ji} = \hat{e}_j \cdot \hat{e}'_i
$$

In order to transform our stresses from one coordinate system to another. 
<p style = "text-align:center;"> 
    
$$
\mathbf{t} = \sigma^T \mathbf{n}
\hspace{2cm} 
\mathbf{t'} = \sigma'^T \mathbf{n'}
$$

Therefore we need to find a relation between $\sigma'$ and $\sigma$. 
<br>
In our new coordinate system: 

$$
\mathbf{t} = \sigma^T \mathbf{n}
\hspace{2cm} 
\mathbf{t'} = \sigma'^T \mathbf{n'}
$$

In our new coordinate system: 

$$
\text{Traction} \hspace{0.1cm}:\hspace{0.1cm} \mathbf{t' = At \Rightarrow t = A^T t'} 
$$

$$
\text{Normal} \hspace{0.1cm}:\hspace{0.1cm}\mathbf{n' = An\Rightarrow n= A^T n'}
$$
    
We also need to remember that transformation matrices are orthogonal. That is: 

$$
\alpha_{ij}^{-1} = \alpha_{ij} \hspace{1cm} (\text{A}^{-1} = \text{A}^\text{T}) $$

Also remember that:

$$  
\alpha_{ij} = \mathbf{\hat{e}'_i \cdot \hat{e}'_j \hspace{2cm} \alpha_{ij}^{-1} =  \hat{e}'_i \cdot \hat{e}'_j = \alpha_{ji} = \alpha_{ij}^T}
$$

So the transformation for our stress tensor is:

 
$$
\mathbf{t' = \color{blue}{A} \sigma^\text{T} n}
$$
    
$$
\mathbf{t' = \color{blue}{A} \sigma^\text{T} \color{red}{A^T} n}
$$

$$
\mathbf{t' =  \sigma^\text{T}  n}
$$

$$
\mathbf{\sigma'^{T} =  \color{blue}{A} \sigma^\text{T} \color{red}{A^T}}
$$
</p>
Using index notation:

$$
\large{\Rightarrow} \hspace{0.5cm} \sigma'^{T}_{ij} = \color{blue}{\alpha_{ik}} \sigma^{T}_{k1} \color{red}{\alpha_{j1}} = \color{blue}{\alpha_{ik}} \color{red}{\alpha_{j1}} \sigma^{T}_{k1}
$$ 
    
Each dependence that the tensor has on direction transforms as a vector, therefore two transformations are required. 

An _n-dimensional_ tensor of _rank r_ consists of $\mathit{n^r}$ components.

This tensor $\text{T}_{i1, i2,...in}$ is defined relative to the basis of the real, linear, n-dimensional space $\mathbf{S}_n$ and under a coordinate transformation $\mathbf{T}$ transforms as: $T'_{ij...n} = \alpha_{ip} \alpha_{jq} \alpha_{nt} \mathbf{T}_{pq...t}$ (i.e one transformation per rank).
<br>
For _orthonormal_ bases, the matrices $\alpha_{ik}$ are _orthogonal_ transformations, i.e $\alpha_{ik}^{-1} = \alpha_{ki}$ (i.e., $\mathbf{A} = \mathbf{A}^T$; columns and rows are orthogonal and have length = 1). 
If the basis is _Cartesian_, $\alpha_{ik}$ are _real_.

## Transforming the 2-D stress tensor  <a name = "2D"></a>

This process lets us determine the normal and shear stress on a plane. 

```{figure} /Chapter2/Images/Lecture2/Fig11.PNG
:width: 40%
```  

We can see above that the force normal to the plane is: 

$$
\mathbf{\hat{n}} = (\sin{\phi}, \cos{\phi})
$$

Therefore the normal stress on the plane is: 

$$
\sigma_{nn} = \mathbf{\hat{n}} \cdot \sigma^T \cdot \mathbf{\hat{n}}
$$

This gives: 

$$
\sigma_{nn} = (\sin{\phi}, \cos{\phi})
\begin{bmatrix}
\alpha_{11} & \alpha_{21} \\
\alpha_{12} & \alpha_{22} \\
\end{bmatrix}
\begin{pmatrix}
\sin{\phi} \\
\cos{\phi} \\
\end{pmatrix}
$$

And similarly the shear stress is given by: 

$$
\sigma_{ns} = \mathbf{\hat{s}} \cdot \sigma^T \cdot \mathbf{\hat{n}} \hspace{0.5cm} \text{where} \hspace{0.5cm} \mathbf{\hat{s}} = (\cos{\phi}, -\sin{\phi})
$$

Mutiplying this out (Verify this yourself) gives:

$$
\sigma_{nn} = \sigma_{11}\sin^2{\phi} + \sigma_{21}\cos{\phi}\sin{\phi} + \sigma_{12} \cos{\phi}\sin{\phi} + \sigma_{22} \cos^2{\phi}
$$

$$
\sigma_{ns} = \sigma_{11}\cos{\phi}\sin{\phi} + \sigma_{21}\cos^2{\phi} - \sigma_{12}\sin^2{\phi} - \sigma_{22} \cos{\phi}\sin{\phi}
$$

This is equivalent to the __tensor transformation__. 

$$
\sigma'^T = \color{blue}{A}\sigma^T\color{red}{A^T}
$$

$$
\sigma'_{qp} = \color{blue}{\alpha_{pi}}\color{red}{\alpha_{qj}}\sigma_{ji}
$$

$$
\sigma'_{nn} = \alpha_{ni}\alpha_{nj}\sigma_{ji}
$$

$$
\sigma'_{ns} = \alpha_{si}\alpha_{nj}\sigma_{ji}
$$

```{figure} /Chapter2/Images/Lecture2/Fig12.png
:width: 75%
``` 

Writing out the full transformation, we know that:

$$
\alpha_{ij} = \mathbf{\hat{e}'_i \cdot \hat{e}_j}
$$

Therefore in the case of a stress tensor: 

$$
\alpha_{s1} = \mathbf{\hat{e}'_s \cdot \hat{e}_1} = \cos{\phi}
$$

$$
\alpha_{s2} = \mathbf{\hat{e}'_s \cdot \hat{e}_2} = -\sin{\phi}
$$

$$
\alpha_{n2} = \mathbf{\hat{e}'_n \cdot \hat{e}_2} = \cos{\phi}
$$

$$
\alpha_{n1} = \mathbf{\hat{e}'_n \cdot \hat{e}_1} = \sin{\phi}
$$

_In tensor notation:_

$$
\sigma'^{T} = \mathbf{A \cdot \sigma^T \cdot A^T}
$$

_In matrix notation:_ 

$$
\begin{bmatrix}
\sigma_{ss} & \sigma_{ns} \\
\sigma_{sn} & \sigma_{nn} \\
\end{bmatrix}
= 
\begin{bmatrix}
\alpha_{s1} & \alpha_{s2} \\
\alpha_{n1} & \alpha_{n2} \\
\end{bmatrix}
\begin{bmatrix}
\alpha_{11} & \alpha_{21} \\
\alpha_{12} & \alpha_{22} \\
\end{bmatrix}
\begin{bmatrix}
\alpha_{s1} & \alpha_{n1} \\
\alpha_{s2} & \alpha_{n2} \\
\end{bmatrix}
$$

$$
\begin{bmatrix}
\sigma_{ss} & \sigma_{ns} \\
\sigma_{sn} & \sigma_{nn} \\
\end{bmatrix}
= 
\begin{bmatrix}
\cos{\phi} & -\sin{\phi} \\
\sin{\phi} & \cos{\phi} \\
\end{bmatrix}
\begin{bmatrix}
\alpha_{11} & \alpha_{21} \\
\alpha_{12} & \alpha_{22} \\
\end{bmatrix}
\begin{bmatrix}
\cos{\phi} & \sin{\phi} \\
-\sin{\phi} & \cos{\phi} \\
\end{bmatrix}
$$

For $\hat{x}_1 = (1,0),\hspace{0.1cm} \hat{x}_2 = (0,1),\hspace{0.1cm} \hat{x}'_1 = (\cos{\phi},-\sin{\phi}) \hspace{0.5cm} \hat{x}'_2 = (\sin{\phi},\cos{\phi})$. 
<br>
The first __row__ of __A__ consists of $\hat{x}'_1$ and the second __row__ consists of $\hat{x}'_2$.

$$
\mathbf{A} = 
\begin{bmatrix}
\mathbf{x'_1\cdot x_1} & \mathbf{x'_1\cdot x_2} \\
\mathbf{x'_2\cdot x_1} & \mathbf{x'_2\cdot x_2} \\
\end{bmatrix}
= 
\begin{bmatrix}
\cos{\phi} & -\sin{\phi} \\
\sin{\phi} & \cos{\phi} \\
\end{bmatrix}
$$

You may recognise $\mathbf{A}$ as a matrix that describes an ___anticlockwise___ rigid-body rotation over an angle $\phi$. 
<br>
$A^T$ describes a ___clockwise___ rotation over an angle $\phi$. The first __column__ of $A^T$ consists of $\hat{x}'_1$ and the second __column__ of $\hat{x}'_2$.


$$
\mathbf{A} = 
\begin{bmatrix}
\mathbf{x_1\cdot x'_1} & \mathbf{x_1\cdot x'_2} \\
\mathbf{x_2\cdot x'_1} & \mathbf{x_2\cdot x'_2} \\
\end{bmatrix}
= 
\begin{bmatrix}
\cos{\phi} & \sin{\phi} \\
-\sin{\phi} & \cos{\phi} \\
\end{bmatrix}
$$

<div class="advanced">

### Diagonalising/Principal Stresses 

Real-valued, symmetric rank 2 tensors (square, symmetric matrices) can only be diagonalised, i.e a coordinate frame can be found, such that only the diagonal elements (normal stresses) remain. 
    
For the stress tensor, these elements $\sigma_1 ,\sigma_2, \sigma_3$ are called the principal stresses.

$$
\begin{bmatrix}
\sigma_1 & 0 & 0 \\
0 & \sigma_2 & 0 \\
0 & 0 & \sigma_3 \\
\end{bmatrix}
$$
    
Such a transformation can be cast as: 
    
$$
\mathbf{T \cdot x = \lambda x}
$$
$$
\Rightarrow \mathbf{(T - \lambda I) \cdot x = 0} 
$$
    
where $\mathbf{x_i}$ are eigenvectors or characteristic vectors and $\lambda_i$ are the eigenvalues, characteristic or principal values. 
    
Note that a non-trivial solution exists only if $\text{det}(\mathbf{T - \lambda I}) = 0$.  

## Stress/Tensor Invariants

Invariants are properties of a tensor that do not change if the coordinate system is changed. 
<br>
A rank-2 tensor has three invariants: 

$$
I_1 = \text{tr}(\mathbf{T}) = T_{11} + T_{22} + T_{33}
$$

$$
I_2 = \text{minor}(\mathbf{T}) =
\begin{vmatrix}
T_{11} & T_{21} \\
T_{21} & T_{22} \\
\end{vmatrix} +
\begin{vmatrix}
T_{11} & T_{31} \\
T_{31} & T_{33} \\
\end{vmatrix} +
\begin{vmatrix}
T_{22} & T_{32} \\
T_{32} & T_{33} \\
\end{vmatrix} 
$$

$$ 
= T_{11} T_{22} + T_{22} T_{33} + T_{11} T_{33} - T_{21}^2 - T_{32}^2 - T_{31}^2 
$$

$$
I_3 = \text{det}(T) = 
\begin{vmatrix}
T_{11} & T_{21} & T_{31} \\
T_{21} & T_{22} & T_{32} \\
T_{13} & T_{23} & T_{33} \\
\end{vmatrix} 
$$

$$
= T_{11} T_{22} T_{33} + 2(T_{21} T_{32} T_{31}) - T_{11}(T_{32})^2 - T_{22}(T_{31})^2 - T_{33}(T_{21})^2 
$$

These invariants are important in operations such as diagonalising. 

## Hydrostatic and Deviatoric Stress 

$$
\sigma = 
\begin{bmatrix}
\sigma_{11} & \sigma_{12} & \sigma_{13} \\
\sigma_{12} & \sigma_{22} & \sigma_{23} \\
\sigma_{12} & \sigma_{23} & \sigma_{33} \\
\end{bmatrix}
$$

$$
\sigma_{ij} = -P\delta_{ij} + \sigma'_{ij}
$$

Each stress element in the stress tensor can be split into two components, the __hydrostatic stress__ and the __deviatoric stress__.

<br>
$ \text{tr}(\sigma) $ is the trace of the stress tensor and is equal to the sum of the normal stresses ( $\sigma_{11} + \sigma_{12} + \sigma_{13}$ ). This is also an invariant of the tensor. 

$$
\text{tr}(\sigma)/3 = \text{ - pressure P = average normal stress = hydrostatic stress}
$$

The __hydrostatic stress__ is the stress that leads to volume change. 

$\sigma_{ij}'$ is the __deviatoric stress__ $ = \sigma_{ij} + P\delta_{ij}$. This is the stress that leads to shape change. 


Note that in geological applications $\sigma_{ij}' \neq \sigma_{ij} + \rho g z \delta_{ij}$
